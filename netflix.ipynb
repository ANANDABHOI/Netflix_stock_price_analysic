```python
df['Date']=pd.to_datetime(df['Date'])
df=df.set_index("Date")
```

```python
df.head()
```

for volume of stock trade


```python
sns.lineplot(x= df.index, y= df.Volume,label="Volume")
plt.title("volume of stock as per date",fontweight="bold")
plt.show()
```

```python
df.plot(y = ["High","Open","Close"])
plt.title("Netflix stock price",fontweight="bold")
```

```python
fig, (ax1,ax2,ax3)=plt.subplots(3,figsize=(15,12))
df.groupby(df.index.day).mean().plot(y="Volume",ax=ax1)
df.groupby(df.index.month).mean().plot(y="Volume",ax=ax2)
df.groupby(df.index.year).mean().plot(y="Volume",ax=ax3)
```

for top 5 dates with highest stock price

```python
a=df.sort_values(by="High",ascending=False).head(5)
a['High']
```

for top 5 dates with low stock price

```python
ab=df.sort_values(by="Low",ascending=False).head(5)
ab['Low']
```

high to low stock price in a periord of time

```python
fig, axes=plt.subplots(nrows=1,ncols=2,sharex=True,figsize=(20,5))
#fig.title("high to low stock price in a periord of time",fontweight="bold")
sns.lineplot(ax=axes[0],y=df["High"],x=df.index ,color="green")
sns.lineplot(ax=axes[1],y=df["Low"],x=df.index ,color="red")
```
